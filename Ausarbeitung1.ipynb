```python
import pandas as pd
```

```python
df = pd.read_csv('results.csv', header=0, index_col='KombinationsID', parse_dates=True)
```

```python
df_aggresiv = df.loc[df['Strategie'] == 'strategies.AggressiveStrategy']
df_defensive = df.loc[df['Strategie'] == 'strategies.DefensiveStrategy']
df_mixed = df.loc[df['Strategie'] == 'strategies.MixedStrategy']
df_movefirst = df.loc[df['Strategie'] == 'strategies.MoveFirstStrategy']
df_movelast = df.loc[df['Strategie'] == 'strategies.MoveLastStrategy']
df_moverandom = df.loc[df['Strategie'] == 'strategies.RandomStrategy']
```

```python
# Import Matplotlib's `pyplot` module as `plt`
import matplotlib.pyplot as plt

# Plot the closing prices for `aapl`
df_aggresiv['Wurde_Gekickt'].plot(grid=True)
df_defensive['Wurde_Gekickt'].plot(grid=True)
df_mixed['Wurde_Gekickt'].plot(grid=True)
df_movefirst['Wurde_Gekickt'].plot(grid=True)
df_movelast['Wurde_Gekickt'].plot(grid=True)
df_moverandom['Wurde_Gekickt'].plot(grid=True)

# Show the plot
plt.show()
```
